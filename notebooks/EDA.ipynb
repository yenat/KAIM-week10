```python
import pandas as pd

# Load the merged data
merged_data = pd.read_csv('../scripts/merged_data.csv')

# Convert 'Date' column to datetime format
merged_data['Date'] = pd.to_datetime(merged_data['Date'], infer_datetime_format=True)

# Display the first few rows of the data
merged_data.head()
```

```python
import matplotlib.pyplot as plt

# Plot the time series data
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['Price'], label='Brent Oil Price')
plt.plot(merged_data['Date'], merged_data['gdp_growth'], label='GDP Growth', linestyle='--')
plt.plot(merged_data['Date'], merged_data['sp500'], label='S&P 500', linestyle=':')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series Data')
plt.legend()
plt.show()
```

```python
# Descriptive statistics
descriptive_stats = merged_data.describe()
descriptive_stats
```

```python
# Manually annotating some significant events
significant_events = {
    '2008-09-15': 'Lehman Brothers Bankruptcy',
    '2020-03-11': 'WHO Declares COVID-19 a Pandemic',
    '2022-02-24': 'Russia Invades Ukraine'
}

# Plot Brent oil prices with significant events annotated
plt.figure(figsize=(12, 6))
plt.plot(merged_data['Date'], merged_data['Price'], label='Brent Oil Price')

# Annotate significant events
for date, event in significant_events.items():
    plt.axvline(pd.to_datetime(date), color='red', linestyle='--', lw=1)
    plt.text(pd.to_datetime(date), merged_data['Price'].max(), event, rotation=90, verticalalignment='center')

plt.xlabel('Date')
plt.ylabel('Brent Oil Price')
plt.title('Brent Oil Price with Significant Events')
plt.legend()
plt.show()
```
